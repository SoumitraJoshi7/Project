```python
import plotly.express as px
# Drop 'Code' and 'Year' columns
country_sum = df.drop(['Code', 'Year', 'Continent'], axis=1)

# Group by 'Country/Territory' and calculate the sum for each cause of death
country_sum = country_sum.groupby(['Country/Territory']).sum()

# Calculate row sums excluding the 'Country/Territory' column
row_sums = country_sum.sum(axis=1)

# Add row sums as a new column in the DataFrame
country_sum['Total Deaths'] = row_sums

country_sum = country_sum.reset_index()
country_total_deaths = country_sum.loc[:,['Country/Territory','Total Deaths']]

#Map
figall = px.choropleth(country_total_deaths,
                    locations='Country/Territory',
                    locationmode='country names',
                    color='Total Deaths',
                    hover_name='Country/Territory',
                    title='Global deaths from 1990 - 2020',
                    color_continuous_scale='Viridis'
                    )

figall.show()
```
